Um tanque de armazenamento contém um líquido à profundidade y. É
tirado líquido a uma vazão constante Q, para atender a demanda, o conteúdo é reposto a
uma taxa senoidal de 3Q sen2(t). A equação que descreve o sistema é dada por,
A.dy = (3Qsen2(t) − Q)dt.
Use um dos métodos estudados para encontrar o instante de tempo em que a profundi-
dade y = 1 m, considerando t ∈ [0 ; 10]. Suponha A = 1200 m2, Q = 500 m3
s e Ea ≤ 10−4.
Solução: xr ≈ 4, 767179. 

```python
import numpy as np
import matplotlib.pyplot as plt

def bissercao(a,b,f,tol):
    if f(a)*f(b) >= 0:
        return print("INTERVALO INVALIDO")
    x0 = 0
    
    while True:
        x = (a+b)/2

        
        if f(a)*f(x) < 0:
            b = x
        
        else:
            a = x

        erro = (x-x0) / x

        if abs(erro) < tol:
            return x

        x0 = x

A = 1200
Q = 500
y = 1.0
        
f = lambda t: (0.5*Q*t - (3*Q *np.sin(2*t))/4) -1200
x = np.linspace(-10,10,100)

plt.plot(x,f(x))
plt.grid()
plt.show()


print(bissercao(0,10,f,10e-7))
```
