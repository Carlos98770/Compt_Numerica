Exercício 6.1 Considere o seguinte funcional de custo
J=
1 n
∑ (yi − f (xi))2
2 i=1
onde f (x) = ax + bsen(x) e (xi ; yi ) é um conjunto de pontos no plano. Para minimizar o
funcional de custo, tome as derivadas parciais em relação as parâmetros a e b, monte o
SEL e encontre sua solução para os pares ordenados:(1; 1, 68), (1, 2; 2, 05) e (2; 2, 51).


```python
x = np.array([1,1.2,2])
y = np.array([1.68,2.05,2.51])

phi = np.array([[sum(x*x),sum(x*np.sin(x))],
                [sum(np.sin(x)*x), sum(np.sin(x)*np.sin(x))]])

f = np.array([[sum(y*x)],[sum(y*np.sin(x))]])

w = np.linalg.inv(phi) @ f

plt.scatter(x,y)
f = lambda x: w[0][0]*x + w[1][0]*np.sin(x)
plt.plot(x,f(x))
plt.show()
print(f"f(x) = {w[0][0]}x + {w[1][0]}sin(x)")
```

```python
phi = np.array([[1,np.sin(1)],[1.2,np.sin(1.2)],[2.51,np.sin(2.51)]])
y = np.array([[1.68],[2.05],[2.51]])



w = np.linalg.inv(phi.T @ phi) @ phi.T@y

z = lambda x: w[0][0]*x+ w[1][0]*np.sin(x)
plt.scatter(x,y)
plt.plot(x,z(x))
print(f"f(x) = {w[0][0]}x + {w[1][0]}sin(x)")
```

Exercício 6.2 Suponha que durante um certo experimento num laboratório de Física,
foi obtido o seguinte conjunto de dados,
t012345
x (t)-0,36982,52986,91628,487113,923814,9146
O pesquisador acredita numa relação linear entre os dados, mas ao fazer uma análise
prévia (por meio do gráfico), não é possível afirmar a existência de uma linha perfeita
devido ao erro de medição. Portanto, se faz necessário uma ajuste nos dados por meio
de algum método. Para isso, iremos considerar o modelo de ajuste x(t) = at + b.


```python
t = np.array([0,1,2,3,4,5])

xt = np.array([-0.36982,2.5298,6.9162,8.4871,13.9238,14.9146])

phi = np.array([[len(t),sum(t)],
                [sum(t),sum(t*t)]])
f = np.array([[sum(xt)],[sum(xt*t)]])



w = np.linalg.inv(phi) @ f

print(w)


f = lambda x: w[0][0] + w[1][0]*x
plt.scatter(t,xt)
plt.plot(t,f(t))
plt.show()

```
