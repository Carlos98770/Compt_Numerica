```python
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = 0.25
c = 0.01
d = e = 0.02
h = 0

#Equa√ß√µes X' = X(a-cX-dY)
         #Y' = -Y(b-eX) - h

x0 = 80
y0 = 20

n = 70

t = np.arange(0,n,0.001)


x = np.zeros(len(t))
y = np.zeros(len(t))

x[0] = x0
y[0] = y0

dt = t[1] - t[0]

for k in range(0,len(t)-1):
  x[k+1] = x[k] + (x[k]*(a- c*x[k] - d*y[k]))*dt
  y[k+1] = y[k] + (-y[k]*(b- e*x[k]) - h)*dt



plt.plot(t, x, label='Presas (x)')
plt.plot(t, y, label='Predadores (y)')
plt.title('Modelo Lotka-Volterra Modificado')
plt.xlabel('Tempo')
plt.ylabel('Popula√ß√£o')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.integrate import odeint


a, b, c, d, e = 1, 0.25, 0.01, 0.02, 0.02


def system(y, t):
    x1, x2 = y
    dydt = [x1 * (a - c * x1 - d * x2),
            -x2 * (b - e * x1)]
    return dydt



y1 = np.linspace(0, 100, 20)
y2 = np.linspace(-5, 120, 20)
Y1, Y2 = np.meshgrid(y1, y2)

t = 0
u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape


for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = system([x, y], t)
        u[i, j] = yprime[0]
        v[i, j] = yprime[1]


Q = plt.quiver(Y1, Y2, u, v, color='r')


plt.xlabel('Presa')
plt.ylabel('Predador')
plt.title('Diagrama de fases')


plt.xlim([0, 100])
plt.ylim([-5, 120])


initial_conditions_list = [[10, 30], [60, 80], [20, 70], [80, 20]]


for initial_conditions in initial_conditions_list:
    t_span = np.linspace(0, 10, 100)
    sol = odeint(system, initial_conditions, t_span)
    plt.plot(sol[:, 0], sol[:, 1], linewidth=2)
    plt.scatter([initial_conditions[0]], [initial_conditions[1]], marker='o')


plt.show()


#Equa√ß√µes X' = X(a-cX-dY)
         #Y' = -Y(b-eX) - h
#Igualando as equa√ß√µes a 0, temos 3 pontos criticos
print()
print(f"Pontos criticos: ({0} , {0}),({100} , {0}), ({25/2} , {4375/8}) ")
```
