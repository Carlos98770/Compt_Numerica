```python
import numpy as np
import matplotlib.pyplot as plt

m1 = 20
m2 = 20
k1 = 1000
k2 = 2000
b1 = 1
b2 = 5
f = 20

Y1_0 = 0
v1_0 = 0
Y2_0 = 1
v2_0 = 0

n = 10

t = np.arange(0,n,0.001)


y1 = np.zeros(len(t))
v1 = np.zeros(len(t))
y2 = np.zeros(len(t))
v2 = np.zeros(len(t))

y1[0] = Y1_0
v1[0] = v1_0
y2[0] = Y2_0
v2[0] = v2_0

dt = t[1] - t[0]


for k in range(0,len(t)-1):
  y1[k+1] = y1[k] + v1[k]*dt
  v1[k+1] = v1[k] + ((-k1*y1[k] - b1*v1[k] + k2*(y2[k] - y1[k]) + b2*(v2[k] - v1[k]))/m1)*dt
  y2[k+1] = y2[k] + v2[k]*dt
  v2[k+1] = v2[k] + ((-k2*(y2[k] - y1[k]) - b2*(v2[k] - v1[k]) + f)/m2)*dt

# Plotting
plt.figure(figsize=(12, 10))

# Plot y1 e y2 vs time
plt.subplot(3, 2, 1)
plt.plot(t, y1, label='y1')
plt.plot(t, y2, label='y2')
plt.title('y1 e y2 vs Tempo')
plt.xlabel('Tempo')
plt.ylabel('Deslocamento')
plt.legend()

# Plot v1 e v2 vs time
plt.subplot(3, 2, 2)
plt.plot(t, v1, label='v1')
plt.plot(t, v2, label='v2')
plt.title('v1 e v2 vs Tempo')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.legend()

# Plot y1 vs v1
plt.subplot(3, 2, 3)
plt.plot(y1, v1)
plt.title('y1 vs v1')
plt.xlabel('y1')
plt.ylabel('v1')

# Plot y2 vs v2
plt.subplot(3, 2, 4)
plt.plot(y2, v2)
plt.title('y2 vs v2')
plt.xlabel('y2')
plt.ylabel('v2')

plt.tight_layout()
plt.show()


```
