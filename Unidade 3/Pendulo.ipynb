```python
import numpy as np
import matplotlib.pyplot as plt

L = 2
L0 = 1
k = 40
m = 1
g = 9.81
theta = (3*np.pi)/4

x1_0 = np.pi/6
x2_0 = 0

n = 10  # Increase the number of time steps

t = np.arange(0, n, 0.001)

x1 = np.zeros(len(t))
x2 = np.zeros(len(t))

x1[0] = x1_0
x2[0] = x2_0

dt = t[1] - t[0]

for i in range(0, len(t)-1):
    x1[i+1] = x1[i] + ((-1/L)*(g*np.sin(theta) + 2*x1[i]*x2[i]))*dt
    x2[i+1] = x2[i] + ((m*L*x1[i]**2 - k*(L-L0) + m*g*np.cos(theta))/m)*dt

plt.plot(t, x1, label="θ(t)")
plt.plot(t,x2,label="ω(t)")
plt.grid()
plt.xlabel("Tempo (s)")
plt.title("Oscilação Amortecida")
plt.legend()
plt.show()


plt.plot(x1,x2)
plt.xlabel("θ(t)")
plt.ylabel("ω(t)")
```

```python
def system(y, t):
    x1, x2 = y
    theta = (3*np.pi)/4
    dydt = [(1 / m) * (m * L * x2**2 * np.sin(theta) + 2 * x1 * x2),
            (m*L*x1**2 - k*(L-L0) + m*g*np.cos(theta))]

    return dydt

y1 = np.linspace(-100, 100, 20)
y2 = np.linspace(-100, 100, 20)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape



for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = system([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]


Q = plt.quiver(Y1, Y2, u, v, color='r')

plt.xlabel('θ(t)')
plt.ylabel('ω(t)')
plt.title('Oscilação Amortecida')
plt.xlim([-100, 100])
plt.ylim([-100, 100])
#plt.savefig('images/retrato_fase.png')
plt.show()
```
